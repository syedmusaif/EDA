- sklearn

- Any sklearn package has 3 steps

    - Step-1: Read the package
    - Step-2: Save the package
    - step-3: Apply fit trasform on data

```python
# Step-1: read the package
from sklearn.preprocessing import LabelEncoder
# step-2: save the package / call the  method
le=LabelEncoder()
# step-3: apply fit transform on data
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
from sklearn.preprocessing import LabelEncoder
visa_df=pd.read_csv(r'C:\Users\syedh\Downloads\Visadataset.csv')
cat_columns=visa_df.select_dtypes(include='object').columns
le=LabelEncoder()
for i in cat_columns:
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df
```

**fit and transform**

- fit means develop the logic

- transform means process the logic

- if we observer in map method first we developed dictionary logic

- the we apply thst dictionary to colum

- Developing the dictionary is kind of fit

- apply the dictionary to columns means we are transfroming the data cst to m=num

- whenever we are transforming  or changing the data then use** tranform

- if we are not changing the data then only fit
